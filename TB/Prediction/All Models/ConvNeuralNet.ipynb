```
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Conv1D, Flatten
```

```
test = pd.read_csv('Final_Test.csv')
train = pd.read_csv("Final_Train.csv")
```

```
final = train.drop(['Pediatric' , 'Geriatric' , '15-24' , '25-34' , '35-44' , '45-54' , '55-64' , 'NEP Total' , 'SumOfHIV_infected'] , axis = 1)
```

```
final['NEP Total'] = train['SumOfHIV_infected']
```

```
final.head()
```

```
final = final.fillna(method ='bfill') 
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(final.values)
scaled_df = pd.DataFrame(scaled , columns = final.columns)
```

```
scaled_df
```

```
from tensorflow.keras.layers import Dropout , MaxPool1D , LSTM
```

```
model = Sequential()
model.add(Conv1D(32, 2, activation="relu", input_shape=(88, 1)))
model.add(Flatten())
model.add(Dense(64, activation="relu"))
model.add(Dense(1))
model.compile(loss="mse", optimizer="adam")
```

```
X = final.drop('NEP Total' , axis = 1)
y = final['NEP Total']
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```
x = X_train.values
```

```
x = x.reshape(2697 , 88 , 1)
```

```
final.shape
```

```
x = X.values.reshape(3372 , 88 , 1)
```

```
model.fit(x, y, batch_size=12,epochs=50, verbose=1)
```
