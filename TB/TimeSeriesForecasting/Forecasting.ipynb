```python
import pandas as pd
import numpy as np
import os
import glob
import pandas as pd
from numpy import array
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
import matplotlib.pyplot as plt
```

```python
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):

        end_ix = i + n_steps

        if end_ix > len(sequence)-1:
            break

        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return array(X), array(y)
```

```python
path = r'./'
all_files = glob.glob(path + "/*.xlsx")
li = []
for filename in all_files:
    df = pd.read_excel(filename, index_col=None, header=0)
    li.append(df)
```

```python
for df in li:
    n_steps = 8
    n_features = 1
    range1 = pd.date_range(start='1/1/2008', periods=44, freq='Q')
    range2 = pd.date_range(start='1/1/2019', periods=24, freq='Q')

    # Pollution

    raw_seq = df['so2']
    X, y = split_sequence(raw_seq, n_steps)
    X = X.reshape((X.shape[0], X.shape[1], n_features))

    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(n_steps, n_features)))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Flatten())
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    model.fit(X, y, epochs=100, verbose=0)

    resArray = list(X[-1])
    for i in range(0 , 24):
      abc = resArray[i:i+8]
      x_input = array(abc).astype('float32')
      x_input = x_input.reshape((1, n_steps, n_features))
      resArray.append(model.predict(x_input, verbose=0))
    output = array(resArray[8:]).flatten()
    result = pd.DataFrame(output , columns = ['so2'])
    result['Year'] = range2

    raw_seq2 = df['no2']
    X2, y2 = split_sequence(raw_seq2, n_steps)
    X2 = X2.reshape((X2.shape[0], X2.shape[1], n_features))
    model.fit(X2, y2, epochs=100, verbose=0)
    resArray2 = list(X2[-1])
    for i in range(0 , 24):
      abc = resArray2[i:i+8]
      x_input = array(abc).astype('float32')
      x_input = x_input.reshape((1, n_steps, n_features))
      resArray2.append(model.predict(x_input, verbose=0))
    output = array(resArray2[8:]).flatten()
    result['no2'] = output

    raw_seq3 = df['rspm']
    X3, y3 = split_sequence(raw_seq3, n_steps)
    X3 = X3.reshape((X3.shape[0], X3.shape[1], n_features))
    model.fit(X3, y3, epochs=100, verbose=0)
    resArray3 = list(X3[-1])
    for i in range(0 , 24):
      abc = resArray3[i:i+8]
      x_input = array(abc).astype('float32')
      x_input = x_input.reshape((1, n_steps, n_features))
      resArray3.append(model.predict(x_input, verbose=0))
    output = array(resArray3[8:]).flatten()
    result['rspm'] = output

    # Climate

    raw_seq = df['rainfall']
    X, y = split_sequence(raw_seq, n_steps)
    X = X.reshape((X.shape[0], X.shape[1], n_features))
    model.compile(optimizer='adam', loss='mse')
    model.fit(X, y, epochs=100, verbose=0)
    resArray = list(X[-1])
    for i in range(0 , 24):
      abc = resArray[i:i+8]
      x_input = array(abc).astype('float32')
      x_input = x_input.reshape((1, n_steps, n_features))
      resArray.append(model.predict(x_input, verbose=0))
    output = array(resArray[8:]).flatten()
    result['Rainfall'] = output

    raw_seq2 = df['relative_humidity']
    X2, y2 = split_sequence(raw_seq2, n_steps)
    X2 = X2.reshape((X2.shape[0], X2.shape[1], n_features))
    model.fit(X2, y2, epochs=100, verbose=0)
    resArray2 = list(X2[-1])
    for i in range(0 , 24):
      abc = resArray2[i:i+8]
      x_input = array(abc).astype('float32')
      x_input = x_input.reshape((1, n_steps, n_features))
      resArray2.append(model.predict(x_input, verbose=0))
    output = array(resArray2[8:]).flatten()
    result['Relative Humidity'] = output
    
    raw_seq3 = df['average_temperature']
    X3, y3 = split_sequence(raw_seq3, n_steps)
    X3 = X3.reshape((X3.shape[0], X3.shape[1], n_features))
    model.fit(X3, y3, epochs=100, verbose=0)
    resArray3 = list(X3[-1])
    for i in range(0 , 24):
      abc = resArray3[i:i+8]
      x_input = array(abc).astype('float32')
      x_input = x_input.reshape((1, n_steps, n_features))
      resArray3.append(model.predict(x_input, verbose=0))
    output = array(resArray3[8:]).flatten()
    result['Average Temperature'] = output

    # TB

    raw_seq = df['NSP Total/ NewMicConf Total']
    X, y = split_sequence(raw_seq, n_steps)
    X = X.reshape((X.shape[0], X.shape[1], n_features))
    model.compile(optimizer='adam', loss='mse')
    model.fit(X, y, epochs=100, verbose=0)
    resArray = list(X[-1])
    for i in range(0 , 24):
      abc = resArray[i:i+8]
      x_input = array(abc).astype('float32')
      x_input = x_input.reshape((1, n_steps, n_features))
      resArray.append(model.predict(x_input, verbose=0))
    output = array(resArray[8:]).flatten()
    result['NSP Total'] = output

    raw_seq2 = df['Relapse Male'] + df['Relapse Female']
    X2, y2 = split_sequence(raw_seq2, n_steps)
    X2 = X2.reshape((X2.shape[0], X2.shape[1], n_features))
    model.fit(X2, y2, epochs=100, verbose=0)
    resArray2 = list(X2[-1])
    for i in range(0 , 24):
      abc = resArray2[i:i+8]
      x_input = array(abc).astype('float32')
      x_input = x_input.reshape((1, n_steps, n_features))
      resArray2.append(model.predict(x_input, verbose=0))
    output = array(resArray2[8:]).flatten()
    result['Relapse Total'] = output

    raw_seq3 = df['NSN Male/NewClinDiag Male'] + df['NSN Female/NewClinDiag Female']
    X3, y3 = split_sequence(raw_seq3, n_steps)
    X3 = X3.reshape((X3.shape[0], X3.shape[1], n_features))
    model.fit(X3, y3, epochs=100, verbose=0)
    resArray3 = list(X3[-1])
    for i in range(0 , 24):
      abc = resArray3[i:i+8]
      x_input = array(abc).astype('float32')
      x_input = x_input.reshape((1, n_steps, n_features))
      resArray3.append(model.predict(x_input, verbose=0))
    output = array(resArray3[8:]).flatten()
    result['NSN Total'] = output

    result['District'] = df.District.unique()[0]
    result.to_csv('./sample_data/Pollution/'+df.District.unique()[0]+'Pred.csv' , index = False)
```
